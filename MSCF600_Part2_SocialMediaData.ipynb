 Part 1: A Deep Dive into the Sources of Social Media DataThe river of social media data is fed by countless tributaries, each with its own unique characteristics, user base, and potential value for financial analysis. Understanding where the data comes from is the first and most critical step in harnessing its power. The information isn’t monolithic; it’s a mosaic of different conversations happening in different digital spaces. Broadly, these sources can be broken down into three main categories: the vast public networks, the focused specialist communities, and the insightful review platforms (Sun et al. 7). Each provides a different piece of the puzzle for understanding market dynamics.Connection-Based Networks: The Public MegaphoneWhen most people think of social media, they think of connection-based networks like X (formerly Twitter), Facebook, and LinkedIn. These platforms are the digital equivalent of a global town square, where hundreds of millions of users—from individual retail traders to CEOs and world leaders—share information openly. Their immense scale and real-time nature make them an invaluable, if noisy, source for gauging broad market sentiment.X, in particular, has become the de facto platform for real-time financial news and commentary. Its structure, which favors brevity and rapid information sharing through "retweets," allows ideas and news to spread with incredible speed. This phenomenon, known as information diffusion, is central to its value. A single influential post can ripple through the network in minutes, causing measurable market impact long before the story hits traditional news outlets. For example, a groundbreaking study by Ge, Kurov, and Wolfe analyzed the market reaction to President Donald Trump's company-specific tweets. They found "significant increases in trading volume and volatility" immediately following a tweet, demonstrating that a single, highly visible source can directly influence investor behavior and asset prices (Ge et al. 215). This highlights the importance of not just the content, but the network structure that allows certain nodes—or influencers—to have an outsized impact on the conversation.Facebook, while less focused on minute-by-minute news, offers a different kind of insight. Its larger user base and more personal nature can provide a clearer picture of general consumer sentiment. Companies often use Facebook to announce new products or engage with customers, and the public reaction within these spaces can be a leading indicator of a product's success or failure. Hasan and Cready found that some firms strategically use Facebook posts to "amplify" their earnings news, suggesting that corporate communication on these platforms is a deliberate part of their investor relations strategy (Hasan and Cready 135). Analyzing the tone and volume of comments on a company's Facebook page can therefore provide clues about consumer confidence and brand loyalty that are difficult to quantify through traditional financial statements alone.Interest-Based Networks: The Specialist's CornerWhile large networks provide breadth, interest-based platforms provide depth. These are the niche communities where users gather to discuss a specific topic, and for finance, the most prominent example is StockTwits. On this platform, every conversation is about the market. Users tag their posts with "cashtags" (e.g., $AAPL for Apple Inc.), creating focused, real-time streams of commentary for specific stocks.The primary value of these platforms lies in the expertise of their user base. The signal-to-noise ratio is often much higher than on a general platform like X because the participants are self-selected investors, traders, and analysts. This concentration of expertise can lead to a phenomenon known as the "wisdom of crowds," where the collective judgment of a group can be surprisingly accurate. However, these platforms are also susceptible to the opposite effect: herding behavior, where investors blindly follow the prevailing trend, sometimes leading to speculative bubbles. Giannini, Irvine, and Shu conducted a fascinating study on StockTwits, analyzing how investor opinions changed around earnings announcements. They found that when investor opinions diverged—that is, when disagreement increased—it was a powerful predictor of higher trading volume and subsequent abnormal returns (Giannini et al. 94). This suggests that the level of consensus or disagreement within these specialist communities is itself a valuable trading signal. The raw sentiment might be positive or negative, but the degree of debate provides a crucial second-layer insight into market uncertainty.Review-Based Networks: The Voice of the Customer and EmployeeThe third key source of social media data comes from review platforms, which offer a ground-level view of a company's operations through the eyes of its most important stakeholders: its customers and its employees. This data is often less about day-to-day market chatter and more about the underlying fundamental health of a business.Customer review sites like Amazon, Yelp, and Trustpilot provide a real-time, high-frequency measure of product popularity and customer satisfaction. A trend of declining star ratings for a company's flagship product can be an early warning sign of quality issues or shifting consumer tastes that may not show up in quarterly sales data for months. Huang's research into Amazon reviews found that "abnormal" review patterns—such as a sudden surge in positive or negative reviews—were predictive of a company's future stock returns and earnings surprises (Huang 164). This indicates that the collective voice of customers contains genuine, forward-looking information about a firm's prospects.Similarly, employee review platforms like Glassdoor and Indeed have opened a window into corporate culture and employee morale. A company with a toxic work environment and high turnover is unlikely to innovate and perform well in the long run. The study by Stamolampros et al. on the hospitality industry found a direct link between job satisfaction ratings on Glassdoor and a firm's financial performance, specifically its return on assets (Stamolampros et al. 130). Green et al. corroborated this, finding that companies with higher employee ratings on Glassdoor tended to outperform their peers in the stock market (Green et al. 236). This data provides a unique angle on a company's intangible assets—its human capital and culture—which are notoriously difficult to assess from a balance sheet alone. By analyzing trends in employee sentiment, investors can gain an edge in identifying well-managed companies poised for long-term success.

### 2. Types of Data

- **Text:** Post content (unstructured).
- **Metadata:** Timestamps, follower counts, engagement metrics.
- **Network:** Follower/retweet graphs and influence pathways.

Each layer supports different analyses: timing (timestamps), influence weighting (followers), and spread (network).

Deconstructing the Anatomy of Social Media DataSocial media data is far more than just a collection of text posts. It is a rich, multi-layered stream of information where the context, the author, and the audience's reaction are just as important as the content itself. To use this data effectively, one must break it down into its core components: the textual content, the surrounding metadata, and the underlying network structure. Each layer provides a different set of signals that can be analyzed to form a holistic view of market sentiment and information flow.Textual Data: Moving Beyond Simple SentimentThe most apparent component of social media data is the text itself—the tweets, posts, comments, and reviews. The initial approach for most analysts is sentiment analysis, which involves classifying the text as positive, negative, or neutral. While this is a crucial first step, it only scratches the surface. To extract deeper insights, analysts employ more sophisticated Natural Language Processing (NLP) techniques.One such technique is topic modeling. Instead of just asking, "Is this post positive or negative?" topic modeling asks, "What is this post about?" Using algorithms like Latent Dirichlet Allocation (LDA), analysts can sift through thousands of posts and identify the key themes or topics of conversation surrounding a particular stock or the market as a whole. For example, during an earnings release, topic modeling could reveal that while overall sentiment is mixed, the dominant conversation among investors is about concerns over "supply chain disruptions" or excitement about a "new product line." This provides a much more granular and actionable insight than a simple sentiment score.Another advanced technique is Named Entity Recognition (NER). NER models are trained to identify and categorize key entities within a text, such as names of people, organizations, and locations. In a financial context, this is incredibly useful for automatically tagging which companies are being discussed in a news article or a social media post, even if they aren't explicitly mentioned with a cashtag like "$AAPL." This allows for a much broader and more accurate collection of relevant data.Furthermore, a significant challenge in analyzing social media text is the prevalence of sarcasm and irony. A post like, "Just love how my portfolio is performing during this 'stable' market," is profoundly negative, but a basic sentiment algorithm might flag the words "love" and "stable" and misclassify it. This has led to the development of specialized sarcasm detection models, which look for contextual clues and linguistic patterns to correctly interpret the true sentiment behind such statements. The complexity of human language in these informal settings means that the quality of textual analysis is directly tied to the sophistication of the NLP models being used.Metadata: The Data About the DataThe metadata accompanying each post provides a wealth of structured information that gives crucial context to the textual content. Ignoring metadata is like reading a headline without knowing the source or the date of publication.Timestamp Analysis is fundamental. The precise time a post was made allows analysts to correlate social media activity with market movements on a tick-by-tick basis. For high-frequency trading firms, a sudden spike in tweet volume or a shift in sentiment that occurs milliseconds before a price jump is a highly valuable signal. For long-term investors, analyzing the evolution of sentiment over days or weeks can reveal shifting narratives and trends. The ability to analyze data at different frequencies makes timestamps one of the most critical pieces of metadata.User Information is another key contextual layer. A bullish post about a little-known biotech stock means one thing if it comes from an anonymous account with ten followers, and something entirely different if it comes from a verified, Nobel Prize-winning biologist with a massive following. Analysts often build models to weigh the importance of a post based on the author's characteristics. This "influencer analysis" can involve metrics like follower count, account age, verification status, and historical accuracy. Some researchers have constructed sophisticated "influencer-weighted sentiment indices," which give more weight to the opinions of users who have proven to be credible or influential in the past. This helps to filter out noise from less informed or spam accounts and focus on the signals that are more likely to be meaningful (Sun et al. 9).Finally, Engagement Metrics—such as likes, shares (retweets), and comment counts—serve as a proxy for the market's attention. A post that gets thousands of retweets is, by definition, a piece of information that a large number of people have deemed important. Some of the most potent signals come from analyzing the rate of engagement. For example, the "retweet velocity"—how quickly a post accumulates retweets in the first few minutes after publication—can be a powerful indicator of the novelty and significance of the information. A high retweet velocity suggests that the information is breaking news that is catching the market's attention in real-time.Network Data: Mapping the Flow of InformationThe third and perhaps most advanced layer of analysis involves looking at the network structure itself—the web of connections between users. This is the domain of graph theory, which models users as "nodes" and their interactions (e.g., following, retweeting, replying) as "edges." By analyzing this graph, we can understand how information flows through the financial community.A key concept here is network centrality. Centrality measures identify the most important or influential nodes within the network. For instance, a user who is retweeted by many other influential users has high centrality and can be considered a key hub of information. Tracking the sentiment of these central nodes can provide a more potent signal than simply averaging the sentiment of all users.Furthermore, network analysis can be used to detect manipulation. Coordinated "pump and dump" schemes often involve a network of seemingly independent accounts that suddenly begin promoting the same stock at the same time. By analyzing the network graph, it's possible to identify these clusters of accounts that are acting in a coordinated fashion, even if they don't follow each other directly. This structural analysis provides a powerful defense against the kind of market manipulation that is rampant on social media. As Grover et al. demonstrated, mapping these online social connections can reveal how corporate narratives are shaped and disseminated, providing a tool to understand both organic conversation and potentially inorganic campaigns (Grover et al. 39).

Part 3: Navigating the Pitfalls of Social media Data Quality

The immense potential of social media data is matched only by its significant quality challenges. It is not a clean, curated dataset like official exchange data; it is a raw, chaotic, and often misleading reflection of human communication. Treating this data as ground truth without rigorous cleaning and critical evaluation is a recipe for disaster. Investors and analysts must be acutely aware of the pervasive issues of bias, noise, and contextual ambiguity that can undermine any model built upon this data.The Problem of Pervasive BiasSocial media data is fundamentally biased because the people creating it are not a representative sample of the world, and their posts are not an unbiased reflection of their true beliefs. This bias manifests in several ways that can systematically skew analysis.Selection Bias is the most obvious issue. The demographics of each social media platform are unique. The user base of TikTok, for example, skews significantly younger than that of Facebook. An analysis of consumer sentiment based on TikTok data might be very predictive for companies targeting Gen Z but utterly useless for those focused on retirees. This demographic skew means that a "market sentiment" index derived from a single platform is not measuring the sentiment of the entire market, but rather the sentiment of that platform's specific user base (Sun et al. 25). Without accounting for this, models can arrive at dangerously flawed conclusions.Another subtle but important issue is Survivorship Bias. In financial discussions online, people are often happy to trumpet their winning trades but quietly delete their losing ones. Over time, this can create a historical record that appears far more prescient than it actually was. An algorithm trained on this skewed history might learn to trust sources that are not actually credible, simply because their track record has been sanitized.Finally, these platforms are often Echo Chambers where users are primarily exposed to information that confirms their existing beliefs. A group of bullish investors on a Reddit forum might spend all day reinforcing each other's positive views, creating a powerful illusion of consensus that is detached from the broader market reality. An analyst who mistakes the sentiment of this echo chamber for the sentiment of the entire market is likely to be caught off guard when the bubble bursts. This is a digital manifestation of confirmation bias, and it makes it difficult to find genuinely new or contrarian information.Noise, Veracity, and the Fight Against ManipulationBeyond bias, the sheer volume of low-quality and intentionally misleading content—the "noise"—is a constant battle. The veracity, or truthfulness, of any given post can never be taken for granted.The Bot Problem is one of the biggest challenges. Sophisticated networks of automated accounts, or "bots," are used for everything from amplifying political messages to executing "pump and dump" schemes in cryptocurrency and penny stock markets. These bots can generate thousands of posts in a short period, creating a completely artificial picture of market sentiment. Identifying these accounts is a complex, ongoing arms race. Researchers use various techniques, from analyzing posting frequency (e.g., an account tweeting 24/7 is likely a bot) to mapping network connections to find coordinated activity. The study by Kraaijeveld and De Smedt, which found that up to 14% of tweets in their crypto dataset were from bots, underscores the scale of this problem (Kraaijeveld and De Smedt). Failure to filter out this bot activity means that any sentiment analysis will be hopelessly contaminated. As Sun et al. note, this noisy and sometimes false information can lead to significant statistical errors, causing models to either find spurious relationships or miss genuine ones (Sun et al. 26).It's also crucial to distinguish between Misinformation (false information spread unintentionally) and Disinformation (false information spread intentionally to deceive). While the former adds to the general noise, the latter represents a direct attack on market integrity. Disinformation campaigns, such as a false rumor about a CEO's health or a fabricated product recall, can be deliberately launched to manipulate a stock's price. Detecting these campaigns requires a combination of textual analysis (e.g., looking for emotionally charged language) and network analysis (e.g., identifying a sudden, coordinated burst of activity from a cluster of accounts).Given these challenges, rigorous Data Cleaning is not optional; it is essential. Practical steps often include filtering out posts from new accounts or accounts with very few followers, removing exact duplicate posts (retweets) to avoid over-weighting popular messages, and applying language complexity filters to screen out nonsensical or spammy content.Contextual Ambiguity: The Nuances of Human LanguageThe final major quality challenge lies in the inherent complexity of human language, especially in the informal context of social media. Models that can't grasp context will constantly make errors.Financial markets have their own rich lexicon of slang and jargon. Terms like "tendies" (gains), "diamond hands" (holding a position despite volatility), or "to the moon" (a belief that a stock will rise dramatically) are opaque to standard NLP models. An effective analysis requires a dictionary or model that has been specifically trained on financial social media language to correctly interpret these terms.Sarcasm and Irony, as mentioned before, are rampant and pose a huge problem for sentiment analysis. The statement, "Another brilliant move by the Fed," could be genuine praise or dripping sarcasm, and the only way to tell is from the broader context of the conversation and the author's history. This ambiguity is a key reason why achieving human-level accuracy in sentiment analysis remains an elusive goal.Finally, the language itself is a moving target. New memes, slang, and emojis emerge constantly. A model trained on data from last year might not understand the meaning of a new emoji or meme that has gone viral this week. This means that NLP models for social media analysis cannot be static; they require continuous monitoring, retraining, and adaptation to keep up with the ever-evolving way that humans communicate online.


### 4. Ethical Issues

 The Ethical Tightrope of Using Social Media DataThe power to extract financial signals from the global conversation comes with profound ethical responsibilities. While the data is often publicly available, its use in high-stakes financial applications raises complex questions about privacy, fairness, and market integrity. Firms that venture into this space without a strong ethical framework not only risk reputational damage but also court legal and regulatory trouble. Navigating this landscape requires a thoughtful approach that goes beyond simply what is technically possible or legally permissible.Privacy and the Question of ConsentThe foundational ethical dilemma revolves around privacy. The dominant argument from data users is that the information is "publicly available"—if someone posts on X, they have no expectation of privacy. However, this argument is increasingly being challenged from an ethical standpoint. There is a significant difference between a person sharing an opinion with their followers and that same opinion being systematically scraped, stored, analyzed, and used to inform billion-dollar trading decisions. The average user has little to no understanding of this secondary use of their data. This creates a disconnect between what users consent to when they use a platform and how their data is actually being used by sophisticated third parties.This issue is further complicated by the limits of data anonymization. While a firm might strip out usernames from a dataset, can a user's entire post history truly be considered anonymous? Often, people share enough personal details over time that their identity can be reverse-engineered, a process known as re-identification. This means that even with precautions, firms may be holding data that is far more sensitive than they realize.Regulators are beginning to step into this gray area. Landmark regulations like the General Data Protection Regulation (GDPR) in Europe and the California Consumer Privacy Act (CCPA) have established stricter rules around the collection and use of personal data. While these laws were not written specifically for financial alternative data, their principles—such as the right to be informed about data collection and the right to have data deleted—are increasingly being applied. A firm that scrapes data from users in these jurisdictions without a clear legal basis for doing so is taking on significant legal risk.Market Manipulation and Systemic RiskThe use of social media data is not just a passive act of observation; it can actively influence and, in some cases, destabilize markets. This creates a set of ethical responsibilities related to market integrity.The most direct threat is market manipulation. The last several years have seen numerous examples of "pump and dump" schemes orchestrated on platforms like Reddit, Discord, and Telegram. In these schemes, a group of individuals collude to artificially inflate the price of a stock (usually a thinly traded penny stock) by creating a sudden burst of coordinated, positive hype online. Once the price has been "pumped," the orchestrators sell their shares to the unsuspecting retail investors who were drawn in by the excitement, causing the price to crash. An investment firm that builds models on social media sentiment has an ethical obligation to ensure its algorithms are not being duped by, and therefore contributing to, these manipulative schemes. This requires building sophisticated detection mechanisms to identify coordinated and inauthentic behavior.There is also a more subtle, systemic risk to consider. As more and more investment firms use similar social media datasets and algorithms, it can create feedback loops in the market. Imagine a scenario where a large number of trading algorithms are all programmed to sell a stock if negative sentiment on X crosses a certain threshold. A sudden, perhaps even artificial, spike in negative sentiment could trigger a wave of automated selling from all these funds at once. This algorithmic selling drives the price down, which in turn generates more real negative sentiment online, triggering even more selling. This is a pro-cyclical feedback loop where the tools designed to read the market end up driving the market in a potentially unstable way. This "crowded trade" phenomenon poses a risk not just to the firms involved, but to the stability of the market as a whole.Algorithmic Bias, Fairness, and TransparencyFinally, there is the critical issue of bias and fairness in the algorithms themselves. The models used to analyze social media data and make trading decisions are not neutral; they reflect the biases inherent in the data they were trained on and the choices made by the people who designed them.Algorithmic bias can manifest in many ways. A sentiment model trained primarily on English-language data from North America will likely perform poorly when analyzing posts from other regions, potentially leading to flawed investment decisions in international markets. Similarly, an algorithm might learn spurious correlations from the data. For example, it might find that stocks frequently discussed on a particular subreddit tend to underperform. If that subreddit is popular with a specific demographic group, the algorithm could end up systematically betting against companies favored by that group, which raises serious questions of fairness.This leads to the "black box" problem. Many of the most powerful models used in this field, such as deep learning neural networks, are incredibly complex. It can be difficult, if not impossible, to understand exactly why the model made a particular decision. If an algorithm makes a trading decision that leads to a huge loss or has a negative societal impact, the firm may be unable to provide a clear explanation for its actions. This lack of transparency and accountability is a major ethical concern. As the use of AI in finance grows, there is an increasing demand from regulators and the public for greater algorithmic accountability—the ability to audit, explain, and take responsibility for the decisions made by automated systems. Firms using social media data have an ethical duty to build their systems in a way that is as transparent and fair as possible, and to be prepared to answer for the consequences of their algorithmic decisions.
Document sources, respect platform TOS, and follow institutional ethics guidance.

**Batch 2 — Exploratory Data Analysis (EDA) and Sentiment Analysis**
🎯 Objective

To transform unstructured social posts into measurable features — specifically sentiment polarity, user influence patterns, and word-level frequency signals — which can later be correlated with market outcomes.

We’ll do this in three analytical layers:

Layer	Focus	Python Tools
L1	Sentiment scoring	nltk.sentiment.vader
L2	Exploratory distribution & grouping	pandas, seaborn
L3	Lexical visualization	WordCloud, matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import nltk

# Theme settings for cleaner plots
plt.style.use('seaborn-v0_8')
sns.set_theme(style="whitegrid")

# Make sure the VADER lexicon is available
try:
    nltk.data.find('sentiment/vader_lexicon.zip')
except LookupError:
    nltk.download('vader_lexicon')
```

**Conceptual Flow**

Attach sentiment to every post using VADER’s compound score.

Summarize behavior — average sentiment per user and per ticker.

Visualize how tone changes over time and across participants.

Generate lexical patterns (WordCloud) to observe topic focus.

This mirrors how hedge-fund data scientists extract alpha signals from noisy conversation data.




**Step 1: Apply Sentiment Analysis**

```python
# Assume df from Batch 1 is already in memory
sid = SentimentIntensityAnalyzer()

# Apply the analyzer row-wise
df['sentiment'] = df['text'].apply(lambda x: sid.polarity_scores(str(x))['compound'])

# Label tone buckets for easier grouping
df['tone'] = pd.cut(df['sentiment'],
                    bins=[-1, -0.05, 0.05, 1],
                    labels=['Negative','Neutral','Positive'])

print("--- Data with Sentiment ---")
print(df[['username','text','sentiment','tone']])
```

**Deep explanation:**

VADER’s compound score condenses multiple lexicon-based features (positive, negative, neutral) into a single metric between −1 (very negative) and +1 (very positive).
We discretize it into tone classes for macro-pattern analysis.

 ### Batch 3 — PCA / SVD / Correlation Layer

**Cell 1 — Imports & small helpers**  

```python
# ======= Cell: Imports & helpers =======
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8')
sns.set_style("whitegrid")
```

**Cell 2 — Construct numeric feature matrix (sentiment, log-followers, TF–IDF for top terms)**

```python
# ======= Cell: Construct numeric feature matrix =======
# Assumes `df` exists from previous cells and includes columns: 'text', 'sentiment', 'followers'
# If you used a different variable name, adapt accordingly.

# 1) Basic numeric features
df_features = df.copy()

# Create a log1p followers column to compress heavy tails (influence scale)
df_features['log_followers'] = np.log1p(df_features['followers'].astype(float))

# Keep sentiment as is (already in [-1,1])
df_features['sentiment_raw'] = df_features['sentiment'].astype(float)

# 2) TF-IDF features: pick top N terms across corpus for manageable dimension
TOP_N_TERMS = 20  # adjust as needed for deeper models; 20 is conservative for demo
vectorizer = TfidfVectorizer(max_features=TOP_N_TERMS,
                             stop_words='english',
                             ngram_range=(1,1),
                             norm='l2')
tfidf = vectorizer.fit_transform(df_features['text'].astype(str))
tfidf_df = pd.DataFrame(tfidf.toarray(), columns=[f"tfidf_{t}" for t in vectorizer.get_feature_names_out()],
                        index=df_features.index)

# 3) Combine numeric columns and TF-IDF into a single DataFrame
numeric_df = df_features[['sentiment_raw','log_followers']].join(tfidf_df)

print("Feature matrix shape:", numeric_df.shape)
numeric_df.head()
```

**Deep notes:**

log1p (log(1 + x)) reduces skew from follower counts while preserving order and zeros.

TF–IDF converts text to numeric importance weights; using max_features=TOP_N_TERMS keeps the feature matrix compact to allow clear PCA interpretation on small samples.

The combined matrix numeric_df is what we’ll analyze with covariance, PCA, and SVD.

**Cell 3 — Standardize features (zero mean, unit variance)**

```python
# ======= Cell: Standardize the numeric matrix =======
scaler = StandardScaler(with_mean=True, with_std=True)
X = scaler.fit_transform(numeric_df.values)  # numpy array (n_samples x n_features)
X_df = pd.DataFrame(X, columns=numeric_df.columns, index=numeric_df.index)

# Quick check
print("Standardized matrix mean (per feature, approx 0):")
print(np.round(X_df.mean(), 6))
print("\nStandardized matrix std (per feature, approx 1):")
print(np.round(X_df.std(ddof=0), 6))
X_df.head()
```

Why standardize?
PCA is sensitive to scale; TF–IDF features and the sentiment/follower columns live on different magnitudes. Standardizing puts all features on equal footing: PCA will then find directions of variance that are not dominated simply by larger-scale features.

**Cell 4 — Covariance and Correlation matrices**

```python
# ======= Cell: Covariance & Correlation =======
cov_matrix = np.cov(X, rowvar=False)  # features on columns => rowvar=False
corr_matrix = np.corrcoef(X, rowvar=False)

cov_df = pd.DataFrame(cov_matrix, index=numeric_df.columns, columns=numeric_df.columns)
corr_df = pd.DataFrame(corr_matrix, index=numeric_df.columns, columns=numeric_df.columns)

print("Covariance matrix (feature x feature) shape:", cov_df.shape)
print("Correlation matrix (feature x feature) shape:", corr_df.shape)

# Display small portion for readability
display(cov_df.round(4))
display(corr_df.round(4))
```

Cell 5 — Visual: Correlation heatmap (features)

```python
# ======= Cell: Correlation heatmap =======
plt.figure(figsize=(10,8))
sns.heatmap(corr_df, cmap='coolwarm', center=0, annot=False, xticklabels=True, yticklabels=True)
plt.title('Feature Correlation Matrix (standardized features)')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

How to read:

Red = positive correlation, blue = negative.

Look for blocks where TF–IDF terms correlate with sentiment or with follower log; those words indicate terms driving mood or attracting influencers.

Interpretation pointers:

Covariance entries show absolute co-variation (units squared). Large covariance between sentiment_raw and a tfidf_term implies the presence of that term correlates with sentiment level.

Correlation (Pearson) standardizes to [-1,1], giving interpretable strength/direction independent of scale.

Cell 6 — PCA: fit, explained variance, scree plot

```python
# ======= Cell: PCA fit & scree plot =======
pca = PCA(n_components=min(X.shape)-0)  # default: all components
pca.fit(X)

explained_var = pca.explained_variance_ratio_  # fraction of variance explained by each PC
cum_explained_var = np.cumsum(explained_var)

# Scree plot & cumulative explained variance
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(range(1, len(explained_var)+1), explained_var*100, color='C0')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance (%)')
plt.title('Scree Plot')

plt.subplot(1,2,2)
plt.plot(range(1, len(cum_explained_var)+1), cum_explained_var*100, marker='o', color='C1')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance (%)')
plt.title('Cumulative Explained Variance')
plt.ylim(0,100)
plt.grid(True)
plt.tight_layout()
plt.show()

# Print top components summary
pc_df = pd.DataFrame({
    'PC': [f'PC{i+1}' for i in range(len(explained_var))],
    'ExplainedVarRatio': explained_var,
    'Cumulative': cum_explained_var
})
display(pc_df.head(10).round(4))
```

**Cell 7 — PCA loadings (eigenvectors) & loadings heatmap**

```python
# ======= Cell: PCA loadings and heatmap =======
loadings = pca.components_  # shape: (n_components, n_features)
loading_df = pd.DataFrame(loadings.T, index=numeric_df.columns,
                          columns=[f'PC{i+1}' for i in range(loadings.shape[0])])

# Show top absolute loadings for first few PCs
num_show = min(6, loading_df.shape[1])
print(f"Top feature loadings (absolute) for first {num_show} PCs:")
for i in range(num_show):
    comp = f'PC{i+1}'
    top_feats = loading_df[comp].abs().sort_values(ascending=False).head(6)
    print(f"\n{comp}:")
    display(loading_df.loc[top_feats.index, comp].round(4))

# Heatmap of loadings (features x first N PCs)
N_PLOT_PCS = min(6, loading_df.shape[1])
plt.figure(figsize=(10,8))
sns.heatmap(loading_df.iloc[:, :N_PLOT_PCS], cmap='vlag', center=0, annot=True, fmt='.2f')
plt.title(f'PCA Loadings (first {N_PLOT_PCS} PCs)')
plt.ylabel('Feature'); plt.xlabel('Principal Components')
plt.tight_layout()
plt.show()
```

What loadings mean (practical):

Each PCA eigenvector (a column in loading_df) gives weights for features that form that principal direction.

Large positive or negative loadings indicate the feature strongly aligns (positive) or anti-aligns (negative) with that PC.

For example, if PC1 has large positive weight on tfidf_earnings and positive on sentiment_raw, PC1 represents a combined “earnings-related positive sentiment” factor.

Cell 8 — Project data onto principal components (scores) and show reconstructed variance

**Interpretation guidance:**

If the first PC explains a large share (say >30–50%), the dataset has a dominant direction (e.g., general sentiment vs. specific-topic variance).

The cumulative plot helps decide how many components to retain for dimensionality reduction (e.g., choose smallest k with cumulative ≥ 80%).

**Cell 9 — Singular Value Decomposition (SVD) on standardized feature matrix and comparison**

```python
# ======= Cell: SVD on standardized matrix and compare to PCA =======
# X is standardized (n_samples x n_features)
U, svals, Vt = np.linalg.svd(X, full_matrices=False)
sing_vals = svals  # singular values
explained_by_svd = (sing_vals**2) / np.sum(sing_vals**2)

# Sanity: PCA explained variance ratios should equal squared singular values normalized
pca_comp_variance = explained_var  # from PCA earlier
svd_variance_equiv = explained_by_svd

# Plot singular values and the variance equivalence
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(range(1, len(sing_vals)+1), sing_vals, 'o-', label='Singular values')
plt.xlabel('Index'); plt.ylabel('Singular value'); plt.title('Singular values (SVD)')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(range(1, len(svd_variance_equiv)+1), np.cumsum(svd_variance_equiv)*100, marker='o', label='SVD cum var')
plt.plot(range(1, len(pca_comp_variance)+1), np.cumsum(pca_comp_variance)*100, marker='x', label='PCA cum var', alpha=0.7)
plt.xlabel('Components'); plt.ylabel('Cumulative explained variance (%)')
plt.title('SVD vs PCA cumulative explained variance')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Print top singular values
print("Top singular values:", np.round(sing_vals[:8], 4))
print("Sum of squared singular values:", np.round(np.sum(sing_vals**2),4))
```

**Key math reminder (why this compares):**

For standardized data X (n×p), PCA on covariance of X returns eigenvalues λ_i. The singular values s_i from SVD on X relate to λ_i by λ_i = (s_i^2)/(n-1) up to conventions (or when X is scaled differently). Practically, the relative proportions (explained variance ratios) from PCA match squared singular values normalized — this is what we plotted.

### Interpretation — What the Covariance, PCA and SVD tell us for this dataset

**Covariance & Correlation:** the covariance matrix measures absolute co-variation between features (TF–IDF term weights, sentiment, follower scale). Large covariances between a word-term and sentiment indicate that when that word appears, sentiment tends to move up (or down). Correlation standardizes those relationships into [-1,1], making them easier to compare.

**PCA (Eigenvectors & Eigenvalues):**
- **Eigenvectors (loadings)** define orthogonal directions in feature space where variance is collected. For example, an eigenvector that loads strongly on `sentiment_raw` and the TF–IDF term 'earnings' indicates a compound axis: posts that mention 'earnings' tend to explain a large share of variance in sentiment. In economic terms, this is a natural 'topic x tone' factor.
- **Eigenvalues** measure how much variance is captured by each eigenvector. A dominant λ1 (first eigenvalue) indicates a single dominant pattern (e.g., general sentiment direction across posts). Smaller eigenvalues correspond to narrower, idiosyncratic patterns (specific words or user-specific signals).

**SVD (Singular Values & Singular Vectors):**
- SVD factorizes the standardized data matrix X into U Σ Vᵀ. The **singular vectors** in V (rows of Vᵀ) are the principal directions (same span as PCA eigenvectors when done on centered/standardized X). The **singular values** in Σ are the square roots of the PCA eigenvalues (up to scaling) and indicate the strength of each orthogonal pattern. In short, SVD gives a numerically stable path to the same decomposition.
- **Why run both?** SVD is numerically more robust on tall/skinny matrices and is the standard computational route in libraries; PCA supplies the statistical interpretation (explained variance) cleanly.

**Practical checks & takeaways for this dataset:**
1. If the first few PCs explain most variance (e.g., 70–90%), we can represent posts with low-dimensional factor scores for downstream tasks (clustering, regression vs. market moves).
2. Examination of loadings helps label components: e.g., PC1 = 'earnings-positive' if it loads on `tfidf_earnings` & sentiment; PC2 might be 'negative technicals' if it loads on 'short' and negative sentiment.
3. Be cautious: TF–IDF sparsity and small sample size can make loadings unstable; use bootstrapping or rolling-window PCA for robustness in production.

**Limitations:**
- PCA & SVD assume linear combinations of features explain most variance — nonlinear patterns (sarcasm, complex semantics) may be missed.
- With small N (few posts), component estimates are noisy. Interpret components as exploratory factors, and validate on larger samples.
